```python
from pandarallel import pandarallel
pandarallel.initialize(progress_bar=False)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
targets = ["target_CHEMBL240",
          "target_CHEMBL260",
          "target_CHEMBL279"]
```

```python
data1 = pd.read_csv("target_CHEMBL240-1.IC50_ready.csv.gz")
data1["target"] = "target_CHEMBL240"

data2 = pd.read_csv("target_CHEMBL260-1.IC50_ready.csv.gz")
data2["target"] = "target_CHEMBL260"

data3 = pd.read_csv("target_CHEMBL279-1.IC50_ready.csv.gz")
data3["target"] = "target_CHEMBL279"                                                
```

```python
# Combine all datasets into a single DataFrame for easier plotting
df = pd.concat([data1, data2, data3])
df
```

```python
import joypy
```

```python
# Set the style
sns.set(style="whitegrid")


# Create the figure with high DPI for a sharper display
plt.figure(figsize=(10, 6), dpi=300)  # Setting dpi here for display purposes

# Create the Joy Plot
fig, axes = joypy.joyplot(df, by='target', column='pchembl_value', 
                          figsize=(10, 5), overlap=3, color=['#1f77b4', '#ff7f0e', '#2ca02c'], 
                          alpha=0.5, linewidth=1)

# Add a title
plt.title('Joy Plot of pChEMBL Values for Three Datasets', fontsize=15)
plt.show()
```
