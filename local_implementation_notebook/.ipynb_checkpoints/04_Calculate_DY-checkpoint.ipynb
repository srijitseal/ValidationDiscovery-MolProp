```python
from sklearn.metrics import mean_squared_error, r2_score

# Load the DataFrame from a CSV file
df = pd.read_csv("output/logp_sorted_fcv_results_RF_testing batches.csv")

fig, axs = plt.subplots(1, 9, figsize=(36, 4), dpi=300)  # Increased figure size

# Iterate over each testing batch and plot in the respective subplot
for i, (batch, group) in enumerate(df.groupby('testing batch')):
    ax = axs[i]
    ax.scatter(group['y_true'], group['y_preds'], color='blue', alpha=0.6, edgecolors='none')

    # Add x=y line
    lims = [
        np.min([ax.get_xlim(), ax.get_ylim()]),  # min of both axes
        np.max([ax.get_xlim(), ax.get_ylim()]),  # max of both axes
    ]
    ax.plot(lims, lims, 'k-', alpha=0.75, zorder=0)
    ax.set_aspect('equal')
    ax.set_xlim(lims)
    ax.set_ylim(lims)

    # Calculate RMSE and R^2
    rmse = np.sqrt(mean_squared_error(group['y_true'], group['y_preds']))
    r_squared = r2_score(group['y_true'], group['y_preds'])

    # Add text for RMSE and R^2
    textstr = f'RMSE: {rmse:.3f}\n$R^2$: {r_squared:.3f}'
    props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
    ax.text(0.05, 0.95, textstr, transform=ax.transAxes, fontsize=12,
            verticalalignment='top', bbox=props)

    ax.set_xlabel('True Values')
    ax.set_ylabel('Predictions')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Adjust layout
plt.show()
```

```python
# Define a function to calculate metrics
def calculate_metrics(group):
    total_compounds = len(group)
    low_error_count = sum((group['y_true'] - group['y_preds']) <= 10)
    discovery_yield = low_error_count / total_compounds if total_compounds > 0 else 0
    return pd.Series({
        'Total Ideal Compounds': total_compounds,
        'Low Error Count': low_error_count,
        'Discovery Yield': discovery_yield
    })

all_results = pd.DataFrame()

# Load the DataFrame from a CSV file
df = pd.read_csv("output/logp_sorted_fcv_results_RF_testing batches.csv")

disc_yld = df[df["y_true"] > 85]

# Group by Testing Batch and apply the function
batch_metrics = disc_yld.groupby('testing batch').apply(calculate_metrics).reset_index()
# Calculate the maximum number of compounds in any batch
max_compounds = batch_metrics['Total Ideal Compounds'].max()

# Standardize the discovery yield
batch_metrics['Standardized Discovery Yield'] = batch_metrics['Discovery Yield'] * (max_compounds / batch_metrics['Total Ideal Compounds'])
batch_metrics['evalt'] = "sorted_fcv"

# Append results
all_results = pd.concat([all_results, batch_metrics], ignore_index=True)


# Optionally, save all results to a single CSV file
all_results.to_csv('metrics/dy_sorted_fcv_discovery_yield_results.csv', index=False)
all_results
```
