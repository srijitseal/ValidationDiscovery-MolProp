```python
import pandas as pd
import numpy as np
```

```python
dy = pd.read_csv('combined_discovery_yield.csv')
dy = dy[dy.Variable!="pchembl"]
dy
```

```python
# Group by and calculate mean and standard deviation
grouped = dy.groupby(["Target", "Method", "evalt"])["Discovery Yield"].agg(['mean']).reset_index()

# Rename the column for clarity
dy = grouped.rename(columns={"mean": "Discovery Yield"})
dy
```

```python
ne = pd.read_csv('combined_novelty_error.csv')
ne = ne.replace({"CV": "cv"}) 
# Group by and calculate mean and standard deviation
grouped = ne.groupby(["Target", "Method", "evalt"])["Novelty Error"].agg(['mean']).reset_index()

# Rename the column for clarity
ne = grouped.rename(columns={"mean": "Novelty Error"})
ne
```

```python
cm = pd.read_csv('combined_metrics.csv')
cm = cm[cm.Sorting_Var!="pchembl"]
cm
```

```python
# Group by and calculate mean and standard deviation
grouped = cm.groupby(["Target", "Method", "evalt"])["R2"].agg(['mean']).reset_index()

# Rename the column for clarity
r2 = grouped.rename(columns={"mean": "R2"})
r2
```

```python
# Group by and calculate mean and standard deviation
grouped = cm.groupby(["Target", "Method", "evalt"])["RMSE"].agg(['mean']).reset_index()

# Rename the column for clarity
RMSE = grouped.rename(columns={"mean": "RMSE"})
RMSE
```

```python
combined_mean = pd.merge(r2, RMSE)
combined_mean = pd.merge(combined_mean, ne)
combined_mean = pd.merge(combined_mean, dy)
```

```python
combined_mean
```

```python
combined_mean = combined_mean[combined_mean["Method"]=="RF"]
combined_mean
```

```python
# Group by 'evalt' and calculate mean and std
grouped_df = combined_mean.groupby("evalt").agg(['mean', 'std']).reset_index()

# Combine mean and std into single columns
grouped_df['R2'] = grouped_df[('R2', 'mean')].round(2).astype(str) + " ± " + grouped_df[('R2', 'std')].round(2).astype(str)
grouped_df['RMSE'] = grouped_df[('RMSE', 'mean')].round(2).astype(str) + " ± " + grouped_df[('RMSE', 'std')].round(2).astype(str)
grouped_df['Novelty Error'] = grouped_df[('Novelty Error', 'mean')].round(2).astype(str) + " ± " + grouped_df[('Novelty Error', 'std')].round(2).astype(str)
grouped_df['Discovery Yield'] = grouped_df[('Discovery Yield', 'mean')].round(2).astype(str) + " ± " + grouped_df[('Discovery Yield', 'std')].round(2).astype(str)

# Select relevant columns
final_df = grouped_df[['evalt', 'R2', 'RMSE', 'Novelty Error', 'Discovery Yield']]
final_df
```
