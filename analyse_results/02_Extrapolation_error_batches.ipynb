```python
import pandas  as pd
import matplotlib.pyplot as plt
```

```python
targets = ["target_CHEMBL240",
          "target_CHEMBL260",
          "target_CHEMBL279"]

methods = ["RF", "xgboost", "MLP"]

sorting_vars = ["logp"]
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def extrapolation_error(df):
    
    df['AbsoluteError'] = abs(df['True Values'] - df['Predictions'])
    # Assign a continuous index to compounds
    df['CompoundIndex'] = range(1, len(df) + 1)

    # Create the scatter plot
    plt.figure(figsize=(10, 6))

    # Scatter plot with continuous x-axis
    sns.scatterplot(x='CompoundIndex', y='AbsoluteError', hue='testing batch', 
                    data=df, palette='Set2', legend=False)

    plt.xlabel('Compounds')
    plt.ylabel('Absolute Error')
    plt.title('Absolute Error of Predictions by Compound')
    plt.tight_layout()
    plt.show()
```

```python
for target in targets:
    for var in sorting_vars:
        for method in methods:
            
            print(target)
            print(method)
            print(var)
            
            df = pd.read_csv(f"../{target}-1_results/{var}_sorted_fcv_results_{method}_testing batches.csv")
            
            extrapolation_error(df)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def extrapolation_error_violin(df):
    # Calculate absolute error
    df['AbsoluteError'] = abs(df['True Values'] - df['Predictions'])
    
    # Create the violin plot
    plt.figure(figsize=(10, 6))
    
    sns.violinplot(x='testing batch', y='AbsoluteError', data=df, palette='Set2')
    
    plt.xlabel('Testing Batch')
    plt.ylabel('Absolute Error')
    plt.title('Absolute Error of Predictions by Batch')
    plt.tight_layout()
    plt.show()

```

```python
for target in targets:
    for var in sorting_vars:
        for method in methods:
            
            print(target)
            print(method)
            print(var)
            
            df = pd.read_csv(f"../{target}-1_results/{var}_sorted_fcv_results_{method}_testing batches.csv")
            
            extrapolation_error_violin(df)
```
