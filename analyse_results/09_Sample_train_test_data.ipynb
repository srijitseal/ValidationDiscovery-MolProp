```python
import pandas  as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
targets = ["target_CHEMBL279"]

methods = ["RF"]

sorting_vars = ["logp"] #, "pchembl"]
```

```python
from pandarallel import pandarallel
pandarallel.initialize()
```

```python

# DataFrame to store all results
results_df = pd.DataFrame(columns=['Target', 'Variable', 'Method', 'Batch', 'evalt', 'Low Similarity Count', 'Novelty Error'])

# Loop through each configuration
for target in targets:
    for var in sorting_vars:
        for method in methods:
            print("sortedFCV")
            print(target)
            print(method)
            print(var)

            # Load the DataFrame from a CSV file
            test_df = pd.read_csv(f"../{target}-1_results/{var}_sorted_fcv_results_{method}_testing batches.csv")
            train_df = pd.read_csv(f"../{target}-1_results/{var}_sorted_fcv_results_{method}_training batches.csv")
            
            break
        break
    break
```

```python
a = train_df[train_df["training batch"]==5]
b = test_df[test_df["testing batch"]==5]
```

```python

# Set the aesthetic style of the plots
sns.set(style="white")
sns.set_context("poster")

# Plotting
plt.figure(figsize=(15, 12), dpi=300)

# Plot histograms
sns.histplot(a['True Values'], label='Training Batch 5', kde=True, alpha=0.6, bins=20)
sns.histplot(b['True Values'], label='Testing Batch 5', kde=True, alpha=0.6, bins=20)

# Add titles and labels
plt.title('Comparison of True Values Distribution for Training and Testing Batch 5')
plt.xlabel('True Values')
plt.ylabel('Frequency')

# Move the legend outside of the plot
plt.legend(title='Dataset', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show plot
plt.show()
```
